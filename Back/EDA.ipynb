```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


plt.style.use('ggplot')
sns.set(style="whitegrid")
%matplotlib inline


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
data = pd.read_csv('data1.csv')
data.head()
```

```python
print(data.info())
data.describe().T
```

```python
target_distribution = data['target'].value_counts(normalize=True) * 100

print("   (%):")
print(target_distribution)

plt.figure(figsize=(6, 4))

sns.countplot(x='target', data=data, palette='Set2', hue='target', legend=False)

plt.title('Distribution of Target Variable')
plt.xlabel('Target')
plt.ylabel('Count')

plt.show()
```

```python
#        
numeric_data = data.select_dtypes(include=[np.number])
mean_by_target = numeric_data.groupby('target').mean().T

mean_by_target.columns = ['Mean for Target=0', 'Mean for Target=1']

mean_by_target['Difference'] = mean_by_target['Mean for Target=1'] - mean_by_target['Mean for Target=0']

mean_by_target.sort_values(by='Difference', ascending=False).head(10)
```

```python
correlation_matrix = data.drop(columns=['ID_code']).corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
# , KDE
def plot_feature_distribution(feature):
    plt.figure(figsize=(10, 4))
    sns.histplot(data[data['target'] == 0][feature], color='blue', label='Target=0', kde=True, stat='density')
    sns.histplot(data[data['target'] == 1][feature], color='orange', label='Target=1', kde=True, stat='density')
    plt.title(f'Distribution of {feature}')
    plt.legend()
    plt.show()

for var in ['var_0', 'var_1', 'var_2']:
    plot_feature_distribution(var)
```

```python
def plot_outliers(feature):
    plt.figure(figsize=(8, 4))
    
    sns.boxplot(x='target', y=feature, data=data, palette='Set2', hue='target', dodge=False, legend=False)
    
    plt.title(f'Boxplot of {feature}')
    plt.show()

for var in ['var_0', 'var_1', 'var_2']:
    plot_outliers(var)
```

```python
from sklearn.feature_selection import f_classif

X = data.drop(columns=['ID_code', 'target'])
y = data['target']
f_scores, p_values = f_classif(X, y)

anova_results = pd.DataFrame({
    'Feature': X.columns,
    'F-Score': f_scores,
    'P-Value': p_values
}).sort_values(by='F-Score', ascending=False)

anova_results.head(10)
```

```python
from sklearn.ensemble import RandomForestClassifier


rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X, y)


feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)


plt.figure(figsize=(10, 6))


sns.barplot(
    x='Importance', 
    y='Feature', 
    data=feature_importances.head(20), 
    palette='viridis', 
    hue=np.ones(len(feature_importances.head(20))), 
    dodge=False, 
    legend=False 
)

plt.title('Top 20 Feature Importances (Random Forest)')
plt.show()
```
